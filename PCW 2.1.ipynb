Let $X$ be a discrete random variable that is the value shown on a single roll of a fair die.

1. Represent the probability density function $f(x)$ in tabular form.

 $\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|c|c|c|c|c|c|c|}
\hline x & 1 & 2 & 3 & 4 & 5 & 6 \T \\\hline
 P(X = x)& 1/6 & 1/6 & 1/6 & 1/6 & 1/6 & 1/6 \T \\\hline
\end{array}$

2. What is the probability that $X=4?$ That $X=4$ or $X=5$?

 $P(X=4)=1/6$
 
 $P(X=4 \text{ or } X=5)=P(X=4)+P(X=5)=1/3$

3. What is the expected value of $X$? Explain the meaning of $E(X)$ in this case.

 The expected value, $E(X)$, is defined as the probabilistically-weighted average of the outcome. This is obtained by summing the products of each possible outcome $x$ with its probability $P(X=x)$. 
 
 $E(X)=\sum_x xP(x) = \frac{1}{6}\sum_{x=1}^6 x = \frac{21}{6}=3.5$

4. Find the expected value of $X^2$

 $E(X^2) = \sum_x x^2 P(x) = \frac {1}{6} \sum_{x=1}^6 x^2 = \frac{90}{6} = 15$

5. Find the variance of $X$.

 $Var(X) = E(X^2)-[E(x)]^2- = 15-3.5^2 = 2.75$ 

6. Obtain a die. Roll it 20 times and record the values obtained. What is the average of the first 5 values? The first 10? What is the average of the 20 rolls?
