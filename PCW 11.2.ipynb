Write a one paragraph summary of Carpenter and Dobkin (2009). Make sure to explain the identification strategy.

Carpenter and Dobkin attempted to determine the causal effect of alcohol consumption on mortality using regression discontinuity. The cutoff point is the minimum legal drinking age of 21, at which point there is expected to be a jump in alcohol consumption, as well as in corresponding mortality. In order to satisfy the zero conditional mean assumption, and reduce the effect of selection bias, the researchers ran a covariate balance test, on the observed covariates of male, white, black, hispanic, HS diploma, employment status and health insurance status. The assumption is that if the covariates of the group slightly before 21 and slightly after 21 are the same, then their expected potential outcomes should be the same and observed differences can be attributed to increase in alcohol consumption.
